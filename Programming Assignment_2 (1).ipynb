1. Write a Python program to convert kilometers to miles?


```python
def Calculate_miles():
    Kilometer = 1000
    miles = Kilometer * 0.621371
    return miles

Calculate_miles()
```

### 2.Write a Python program to convert Celsius to Fahrenheit?

```python
def Calculate_fahrenheit(cs):
    Celsius = cs
    fahrenheit = (Celsius * 9/5) + 32
    return fahrenheit

Calculate_fahrenheit(1)
```

3. Write a Python program to display calendar?


```python
import calendar

def display_calendar():
    Cal = calendar.monthcalendar(2023,5)
    calendar_text = calendar.month_name[5] + ' ' + str(2023)
    calendar_text += '\n' + 'Mon Tue Wed Thur Fri Sat Sun \n'
    for week in Cal:
        for day in week:
            if day != 0:
                calendar_text += f' {day:2d} '
            else:
                calendar_text += '  '
        calendar_text += '\n'
    print(calendar_text)

display_calendar()
```

### 4. Write a Python program to solve quadratic equation?

```python
import cmath
def Quadratic_equation(a,b,c):

    dics = (b**2) - (4 * a * c)

    root1 = (-b + cmath.sqrt(dics)) / (2 * a)
    root2 = (-b - cmath.sqrt(dics)) / (2 * a)
    
    return root1 , root2 

root1 , root2 = Quadratic_equation(1 , -5, 10)
print(f"The roots are: {root1} and {root2}")
```
